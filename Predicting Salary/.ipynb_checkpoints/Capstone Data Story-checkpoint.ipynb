## Capstone Project data story - predicting big incoming transactions
### How I retrieved the data
I extracted the data from a table containing all transactions in a Netezza database (IBM). At first I will be doing the analyses on a laptop, that's why I also extracted a 0.1%-sample from the transaction table. 

I used this query to retrieve 0.1%-the sample:

*drop table fmbtd_01_perc;*  

*create table fmbtd_01_perc as*  
*select cast (cast (fm.TIJDID as char(8)) as date) as date, fm.tijdid, fm.relnr, fm.ovkid as arrnr, fm.lftid as age , fm.bijbd*   *as amount, fm.tegreknr as counterpartyaccountnr, koz.LEEFTIJD as age_201601, koz.GESLACHT as gender, koz.DREL from*   *p_ori_w..FMBTD fm*  
*left join p_ori_w..fkozmos koz*  
*on fm.RELNR = koz.RELNR*  
*where fm.bijbd is not null*  
*and mod (fm.relnr, 1000) =251*  
*and koz.TIJDID = 20160100*  
*;*

## Predicting big incoming regular transactions
**Still, we would like to help customers gain more insight in their financial situation, to support customers in their decion making process. Therefore we need to find a way to predict the biggest incoming regular transaction. **
To do so, let's investigate some specific customers and try to find an initial algorithm for finding the biggest regular incoming transaction.

```python
# Create a set of 5 relnrs to gain some insight in customer transactions
relnr_list = [31873251, 8904251, 17808251, 15042251, 58708251, 109338251]
```

```python
cust_trx = trx[trx.relnr.isin(relnr_list)].sort_index()
```

```python
fig, axes = plt.subplots(nrows =3, ncols = 2, figsize=(16,18))
fig.tight_layout()
plt.subplots_adjust(hspace=0.3)

titles = []
for i in range(3):
    for j in range(2):
        titles.append('Transactions of customer %s' % relnr_list[i*2+j])
    
        # Remove the plot frame lines. They are unnecessary.  
        axes[i,j].spines["top"].set_visible(False)  
        axes[i,j].spines["right"].set_visible(False)  
  
        # Ensure that the axis ticks only show up on the bottom and left of the plot.  
        # Ticks on the right and top of the plot are generally unnecessary chartjunk.  
        axes[i,j].get_xaxis().tick_bottom()  
        axes[i,j].get_yaxis().tick_left() 
        axes[i,j].set_title(titles[i*2+j])
        axes[i,j].set_ylabel('Amount')
        
        cust_trx[cust_trx.relnr == relnr_list[i*2+j]].amount.plot(ax = axes[i,j], color = '#525199')    
```

All but the last one seem to have a regular big incoming transaction, but they almost all have big outliers. These can sometimes be explained due to vacation money (May) and/or the thirteenth month (January). In the last one almost only small amounts are coming into the account. Let's remove these outliers and look at the remaining transactions. 

```python
fig, axes = plt.subplots(nrows =3, ncols = 2, figsize=(16,18))
fig.tight_layout()
plt.subplots_adjust(hspace=0.3)
limits = [7000, 3500, 1500, 2500, 2500, 600]

titles = []
for i in range(3):
    for j in range(2):
        titles.append('Transactions of customer %s' % relnr_list[i*2+j])
    
        # Remove the plot frame lines. They are unnecessary.  
        axes[i,j].spines["top"].set_visible(False)  
        axes[i,j].spines["right"].set_visible(False)  
  
        # Ensure that the axis ticks only show up on the bottom and left of the plot.  
        # Ticks on the right and top of the plot are generally unnecessary chartjunk.  
        axes[i,j].get_xaxis().tick_bottom()  
        axes[i,j].get_yaxis().tick_left() 
        axes[i,j].set_title(titles[i*2+j])
        axes[i,j].set_ylabel('Amount')
        
        cust_trx[(cust_trx.relnr == relnr_list[i*2+j]) & (cust_trx.amount <= limits[i*2+j])].amount.plot(ax = axes[i,j], color = '#FF6200')    
```

The last one remains difficult when looking at regular incoming transaction. Nevertheless, the others definitely seem to have. Let's check out the the individual transactions and where they're coming from. For that we can check the counterpartyaccountnr.

```python
 cust_trx[(cust_trx.relnr == relnr_list[0]) & (cust_trx.amount >= 4000) & (cust_trx.amount <= 7000)].counterpartyaccountnr.value_counts()
```

```python
cust_trx[(cust_trx.relnr == relnr_list[1]) & (cust_trx.amount >= 1500) & (cust_trx.amount <= 3500)].counterpartyaccountnr.value_counts()
```

```python
cust_trx[(cust_trx.relnr == relnr_list[2]) & (cust_trx.amount >= 800) & (cust_trx.amount <= 1500)].counterpartyaccountnr.value_counts()
```

```python
cust_trx[(cust_trx.relnr == relnr_list[3]) & (cust_trx.amount >= 1500) & (cust_trx.amount <= 2500)].counterpartyaccountnr.value_counts()
```

```python
cust_trx[(cust_trx.relnr == relnr_list[4]) & (cust_trx.amount >= 1500) & (cust_trx.amount <= 2500)].counterpartyaccountnr.value_counts()
```

So, actually in all cases you can detect a transaction that happens often from the same counterparty. In the second case I 
suspect a change of jobs or a change in accountnumber of the employer. Also, notice that counterpartyaccountnumber 769500 
appears twice, could have something to do with taxes. Interesting, so the next step will be looking at how regular the transactions are.
